# Explorarity Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path="D:/Untar/Project/Tableau/Income Classification/Income_Evaluation_V2.csv"
ice= pd.read_csv(path, skipinitialspace=True, index_col=0)
ice.head()
```

## Generation

```python
colors = sns.color_palette('pastel')[0:5]
generation = ice['generation'].value_counts()
plt.figure(figsize = (10, 5))
plt.pie(generation, labels = generation.index, colors= colors, autopct = '%1.2f%%')
plt.title('generation')
plt.show()
```

## Workclass

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='workclass', data = ice)
plt.show()
```
